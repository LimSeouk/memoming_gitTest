```python
# (001)정수기
# placeholder 만들기
X_001 = tf.placeholder(tf.float32, shape=[None,5])
Y_001 = tf.placeholder(tf.float32, shape=[None,1])
# 가중치값 초기화
W_001 = tf.Variable(tf.random_normal([5,1],name='weight'))
# 바이어스값 초기화
b_001 = tf.Variable(tf.random_normal([1]), name='bias')

# 가설식 설정
hypothesis_001 = tf.matmul(X_001,W_001)+b_001
# 비용함수 설정
cost_001 = tf.reduce_mean(tf.square(hypothesis_001-Y_001))
# 최적화함수 설정
optimizer_001 = tf.train.GradientDescentOptimizer(learning_rate=0.000005)
train_001 = optimizer_001.minimize(cost_001)

# 세션 초기화
sess_001 = tf.Session()
# 글로벌 변수 초기화
sess_001.run(tf.global_variables_initializer())
# 학습진행
for step in range(300000) :
    cost_, hypo_, _ = sess_001.run([cost_001,hypothesis_001,train_001],feed_dict={X_001:x_data_001,Y_001:y_data_001})
    if step % 10000 == 0:
        print("#",step," 손실 비용:",cost_)
        print("- 판매량 : ", hypo_[0])
```
