```python
from sklearn.cluster import KMeans
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import seaborn as sb
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame(columns=['x','y'])
```

```python
df.loc[0] =[8,15]
df.loc[1] =[10,10]
df.loc[2] =[15,15]
df.loc[3] =[19,8]
df.loc[4] =[6,4]
df.loc[5] =[8,17]
df.loc[6] =[8,20]
df.loc[7] =[19,5]
df.loc[8] =[18,10]
df.loc[9] =[1,2]
df.loc[10] =[10,20]
df.loc[11] =[11,19]
df.loc[12] =[8,12]
df.loc[13] =[3,8]
df.loc[14] =[11,3]
df.loc[15] =[6,12]
df.loc[16] =[9,15]
df.loc[17] =[17,11]
df.loc[18] =[12,2]
df.loc[19] =[9,14]
df.loc[20] =[12,13]
df.loc[21] =[17,20]
df.loc[22] =[5,11]
df.loc[23] =[18,20]
df.loc[24] =[7,6]
df.loc[25] =[11,2]
df.loc[26] =[10,9]
df.loc[27] =[3,17]
df.loc[28] =[13,14]
df.loc[29] =[17,15]
```

```python
df.head(30)
```

```python
sb.lmplot('x','y', data=df, fit_reg=False, scatter_kws={"s":100})
plt.title('K-Means Example')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# points라는 변수를 만들어서 numpy 객체를 초기화
points = df.values
# kmeans알고리즘을 수행해서 클러스터 4개를 만들도록 수행
kmeans = KMeans(n_clusters=4).fit(points)
# 각클러스터의 중심위치를 구할 수 있도록 만듬
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
df['cluster'] = kmeans.labels_
df.head(30)
```

```python
sb.lmplot('x','y', data=df, fit_reg=False, scatter_kws={'s':150},hue='cluster')
plt.title('K-Means Example')
plt.xlabel('x')
plt.ylabel('y')
```
